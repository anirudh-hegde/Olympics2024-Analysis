```python
df=df[df['height']!=0.0]
name=df['name'].head(10)
height=df['height'].head(10)
plt.xticks(rotation=90)
sns.lineplot(x=name,y=height,color='green',marker='o',markersize=10)
plt.xlabel('Name')
```

```python
df=df[df['lang']!=""]
```

```python

plt.bar(df['lang'].head(20).value_counts().index,df['lang'].head(20).value_counts())
plt.xlabel('Language')
plt.xticks(rotation=90)
plt.ylabel('Number of athletes')
plt.show()
```

```python
sns.countplot(df['lang'].head(20).value_counts(),palette='prism')
plt.xlabel('Language')
plt.xticks(rotation=90)
plt.ylabel('Number of athletes')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))


sns.countplot(x='lang', data=df, ax=axes[0, 0], palette='prism')
axes[0, 0].set_title('Number of Athletes by Language')
axes[0, 0].set_xlabel('Language')
axes[0, 0].set_ylabel('Number of Athletes')
axes[0, 0].tick_params(axis='x', rotation=90)


sns.histplot(df['height'], kde=True, ax=axes[0, 1], color='green')
axes[0, 1].set_title('Height Distribution of Athletes')
axes[0, 1].set_xlabel('Height')
axes[0, 1].set_ylabel('Frequency')


sns.countplot(x='residence_country', data=df, ax=axes[1, 0], palette='coolwarm')
axes[1, 0].set_title('Number of Athletes by Country')
axes[1, 0].set_xlabel('Country')
axes[1, 0].set_ylabel('Number of Athletes')
axes[1, 0].tick_params(axis='x', rotation=90)


top_10 = df.nlargest(10, 'height')
sns.lineplot(x=top_10['name'], y=top_10['height'], ax=axes[1, 1], marker='o', color='purple')
axes[1, 1].set_title('Top 10 Tallest Athletes')
axes[1, 1].set_xlabel('Name')
axes[1, 1].set_ylabel('Height')
axes[1, 1].tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```
